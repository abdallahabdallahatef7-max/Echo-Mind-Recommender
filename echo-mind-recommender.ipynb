```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df=pd.read_csv("/kaggle/input/-spotify-tracks-dataset/dataset.csv")
df
```

```python
df = df.drop_duplicates(subset=['track_name', 'artists']).dropna()
```

```python
features = ['danceability', 'energy', 'loudness', 'speechiness', 
            'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']
```

```python
scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(df[features]).astype('float32')
```

```python
def get_recommendations(song_name, n=5):
    try:
        
        idx = df[df['track_name'].str.lower() == song_name.lower()].index[0]
        song_vector = df_scaled[idx].reshape(1, -1)
        
        
        scores = cosine_similarity(song_vector, df_scaled)[0]
        
       
        similar_indices = scores.argsort()[-(n+1):-1][::-1]
        
        return df.iloc[similar_indices][['track_name', 'artists', 'album_name']]
    except:
        return "Song not found!"
```
